```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
os.listdir("./data")
```

```python
barbers = pd.read_csv('./data/barbers.csv')
longshore = pd.read_csv('./data/longshore.csv')
building = pd.read_csv('./data/table 1 building trades.csv')
clothing = pd.read_csv('./data/table 2 clothing and garment trades.csv')
food = pd.read_csv('./data/table 3 food and beverages.csv')
```

```python
longshore.head(10)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set(font_scale=2)
sns.set_palette("Paired")
ax = sns.lineplot(data=longshore, x="year", y="wage_adj", hue="Occupation")
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
ax.set_title("Boston Massachusetts Longshoremen Wages 1928-1933", size=30)
ax.set_ylabel("Wages in Dollars per hour")
fig = ax.get_figure()
fig.savefig("./figures/longshore.png", bbox_inches="tight") 
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.set(font_scale=2)
sns.set_palette("Paired")
ax = sns.lineplot(data=barbers, x="Year", y="Barbers_wage_adj")
# ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
ax.set_title("Boston Massachusetts Barbers Wages 1928-1933", size=30)
ax.set_ylabel("Wages in Dollars per hour")
fig = ax.get_figure()
fig.savefig("./figures/barbers.png") 
```
